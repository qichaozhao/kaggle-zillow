```python
df = pd.read_csv('data/raw/properties_2016.csv')
```

```python
# Let's do some exploratory data analysis on this set
# First lets see how many NA values we have per column
```

```python
fig = plt.figure(figsize=(20,20))
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
# There are a lot of columns with na values - lets drop all columns which only have NA values
# We count it as droppable if it has more than 25% NA values
ddf = df.dropna(how='all', axis=1, thresh=(len(df) * 0.75))
```

```python
ddf.to_csv('data/interim/na_cleaned.csv.gz', compression='gzip', index=False)
```
